```python
# Define the Brazilian IPCA inflation rates for 2001-2021
inflation_rates = {
    2001: 0.0765,
    2002: 0.1253,
    2003: 0.0976,
    2004: 0.0706,
    2005: 0.0609,
    2006: 0.0366,
    2007: 0.0452,
    2008: 0.0593,
    2009: 0.0474,
    2010: 0.0514,
    2011: 0.0681,
    2012: 0.0582,
    2013: 0.0591,
    2014: 0.0637,
    2015: 0.1082,
    2016: 0.0629,
    2017: 0.0294,
    2018: 0.0353,
    2019: 0.0433,
    2020: 0.0412,
    2021: 0.1026
}

# Function to adjust for inflation
def adjust_for_inflation(value, start_year, end_year):
    inflation_adjusted_value = value
    for year in range(start_year, end_year):
        inflation_adjusted_value /= (1 + inflation_rates[year])
    return inflation_adjusted_value
```

```python

years = list(range(2001, 2023))

# Adjust the exported values for inflation
soy_adjusted = [adjust_for_inflation(value, year, 2022) for year, value in zip(years, soy_data.loc[:, 'Exported value in 2001':'Exported value in 2022'].values[0])]
lumber_adjusted = [adjust_for_inflation(value, year, 2022) for year, value in zip(years, lumber_data.loc[:, 'Exported value in 2001':'Exported value in 2022'].values[0])]
beef_adjusted = [adjust_for_inflation(value, year, 2022) for year, value in zip(years, meat_data.loc[:, 'Exported value in 2001':'Exported value in 2022'].values[0])]

# Create the merged DataFrame
merged_dft = pd.DataFrame({'Year': years,
                           'Soy Exported Value': soy_adjusted,
                           'Lumber Exported Value': lumber_adjusted,
                           'Beef Exported Value': beef_adjusted})

print(merged_dft)
```

```python
correlation_matrix = yoy_change1.corr()

# Display the correlation matrix
# print(correlation_matrix)

# correlation_matrix.to_excel('Correlation Price Output.xlsx', index=False)

```

```python
from scipy.stats import pearsonr
# Determine the dimensions of the matrix
rows, cols = correlation_matrix.shape
# Initialize an empty matrix for storing p-values
p_values = np.empty((rows, cols))

# Calculate p-values for each pair of variables
for i in range(rows):
    for j in range(cols):
        if i == j:
            p_values[i, j] = np.nan  # Diagonal elements are not compared
        else:
            p_values[i, j] = pearsonr(correlation_matrix.iloc[:, i], correlation_matrix.iloc[:, j])[1]

# # Print p-values matrix
# print("P-values matrix:")
# print(p_values)

# Define the significance level (alpha) for the test
alpha = 0.05

# Check which correlations are statistically significant
significant_corr = p_values < alpha

## Convert significant_corr to a DataFrame
significant_corr_df = pd.DataFrame(
    significant_corr, index=correlation_matrix.columns, columns=correlation_matrix.columns
)

# Save the DataFrame to an Excel file
significant_corr_df.to_excel('significant_corr.xlsx')


# # Print the significant correlations matrix
# print("Significant correlations matrix:")
print(significant_corr)


# Convert p_values to a DataFrame
p_values_df = pd.DataFrame(p_values, index=correlation_matrix.columns, columns=correlation_matrix.columns)

# Save the DataFrame to an Excel file
p_values_df.to_excel('p_values.xlsx')


```

```python
import pandas as pd

yoy_corr = pd.read_excel("/Users/johncrye/Desktop/Capstone_Data_Local_File/Correlation Price Output.xlsx")
yoy_corr_sig = pd.read_excel("/Users/johncrye/Desktop/Capstone_Data_Local_File/significant_corr.xlsx")

def print_values(df1, df2):
    results = []
    for i in range(len(df1)):
        for j in range(1, len(df1.columns)):
            results.append((df1.index[i], df1.columns[j], df1.iloc[i, j], df2.iloc[i, j]))
    return results

yoy_corr.set_index(yoy_corr.columns[0], inplace=True)
yoy_corr_sig.set_index(yoy_corr_sig.columns[0], inplace=True)

results = print_values(yoy_corr, yoy_corr_sig)

# Separate significant and non-significant correlations
significant = [(x[0], x[1], x[2]) for x in results if x[3]]
not_significant = [(x[0], x[1], x[2]) for x in results if not x[3]]

# Create a DataFrame with four columns: Significant terms, Significant values, Non-Significant terms, Non-Significant values
output_df = pd.DataFrame({"Significant Terms": pd.Series([f"{x[0]}, {x[1]}" for x in significant]),
                          "Significant Values": pd.Series([x[2] for x in significant]),
                          "Non-Significant Terms": pd.Series([f"{x[0]}, {x[1]}" for x in not_significant]),
                          "Non-Significant Values": pd.Series([x[2] for x in not_significant])})

# Export the DataFrame to an Excel file
output_df.to_excel("correlation_results_4_columns.xlsx", index=False)

# Create a new DataFrame for export with index, column value, correlation value, and true or false as columns
export_df = pd.DataFrame({"Index": pd.Series([x[0] for x in results]),
                          "Column": pd.Series([x[1] for x in results]),
                          "Correlation Value": pd.Series([x[2] for x in results]),
                          "Significant": pd.Series([x[3] for x in results])})

# Export the new DataFrame to an Excel file
export_df.to_excel("correlation_results_full_info.xlsx", index=False)
```

```python
def remove_highly_correlated_variables(df, threshold):
    # Calculate the correlation matrix
    corr_matrix = df.corr().abs()

    # Find highly correlated variable pairs
    highly_correlated_pairs = [(i, j) for i in range(len(corr_matrix.columns)) for j in range(i+1, len(corr_matrix.columns)) if corr_matrix.iloc[i, j] >= threshold]

    # Remove one of the variables in each highly correlated pair
    while highly_correlated_pairs:
        i, j = highly_correlated_pairs.pop()
        col_to_remove = corr_matrix.columns[j]

        # Drop the highly correlated column
        df = df.drop(col_to_remove, axis=1)

        # Recalculate the correlation matrix
        corr_matrix = df.corr().abs()

        # Update the list of highly correlated pairs
        highly_correlated_pairs = [(i, j) for i in range(len(corr_matrix.columns)) for j in range(i+1, len(corr_matrix.columns)) if corr_matrix.iloc[i, j] >= threshold]

    return df

# Sample usage
threshold = 0.8
data = pd.read_excel('correlation_results_full_info.xlsx')
 # Load your data into a DataFrame
filtered_data = remove_highly_correlated_variables(data, threshold)


data.to_excel('correlation_results_filtered.xlsx', index=False)

threshold = 0.8
data = pd.read_excel('correlation_results_filtered.xlsx')
filtered_data = remove_highly_correlated_variables(data, threshold)
```
