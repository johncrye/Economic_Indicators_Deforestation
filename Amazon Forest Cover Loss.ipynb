```python
data = data.reset_index()
data.columns = ["DATE", "PBEEFUSDM"]
data["DATE"] = pd.to_datetime(data["DATE"])
```

```python
# create a new column for the year
data["YEAR"] = data["DATE"].dt.year

# group the data by year and calculate the median, min, and max for each year
yearly_data = data.groupby("YEAR").agg({"PBEEFUSDM": ["min", "max", "median"]})

# flatten the column names of the resulting DataFrame
yearly_data.columns = [f"{col[0]}_{col[1].upper()}" for col in yearly_data.columns]

# reset the index to make "YEAR" a column again
yearly_data = yearly_data.reset_index()

# print the resulting DataFrame
print(yearly_data)
```

```python
import matplotlib.pyplot as plt

# create a line chart of median prices over time
plt.plot(data['YEAR'], data['PBEEFUSDM_MEDIAN'], label='Median Price')

# add error bars to indicate the range of minimum and maximum prices
plt.errorbar(data['YEAR'], data['PBEEFUSDM_MEDIAN'], 
             yerr=[data['PBEEFUSDM_MEDIAN'] - data['PBEEFUSDM_MIN'], 
                   data['PBEEFUSDM_MAX'] - data['PBEEFUSDM_MEDIAN']], 
             fmt='none', ecolor='gray', capsize=2)

plt.xlabel('Year')
plt.ylabel('Price (USD)')
plt.title('Beef Price Range over Time')
plt.legend()
plt.show()
```
