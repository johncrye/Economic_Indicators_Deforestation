```python
df = pd.read_csv("SeriesExport-11-12-2022_15-08-33.csv")
```

```python
df.head(10)
```

```python
column_names = list(df.columns)
filtered_columns = [name for name in column_names if not "Series Name." in name]
# filtered_columns
df = df[filtered_columns]
df.head(10)
df = df.drop(range(6)).reset_index(drop=True)
```

```python
df
```

```python
Biomass_columns = [name for name in filtered_columns if ("Biomass" in name) or ("Series Name" in name)]
# filtered_columns
# print(Biomass_columns)
Biomass_df = df[Biomass_columns]
Biomass_df = Biomass_df.dropna()
Biomass_df
```

```python
def find_non_ints(row):
    for i in range(len(row)):
#         print(row[i])
        if "(" in str(row[i]) or "ie" in str(row[i]):
#             print("here")
            row[i] = 0.0
        else:
#             print(row[i])
            row[i] = float(row[i])
            
#     print("here")
    return row
    
Biomass_df = Biomass_df.apply(lambda row: find_non_ints(row), axis=1)
Biomass_df
```

```python
list(Biomass_df.columns)[1].split(",")[1].strip()
```

```python
cleaned_column_names = [column.split(",")[1].strip() for column in list(Biomass_df.columns)[1:]]
print(["Series Name"]+cleaned_column_names)
Biomass_df.columns = ["Series Name"] + cleaned_column_names
lines = Biomass_df.plot.line(x="Series Name", title = "Biomass Electricity Generation in Billion kWh")
```

```python
# select the relevant columns from your DataFrame
df2 = Biomass_df[['Series Name', 'Paraguay', 'Suriname', 'Argentina', 'Peru', 'Ecuador', 'Brazil', 'Uruguay', 'Chile', 'Colombia', 'Venezuela', 'French Guiana', 'Guyana']]

# create a new column that contains the sum of biomass electricity generation for all the countries/regions except Brazil
df2['Rest of series'] = df2.sum(axis=1) - df2['Brazil'] - df2['Series Name']

print(df2)
```

```python

# create a line graph
ax = df2.plot(kind='line', x='Series Name', y=['Brazil', 'Rest of series'], title='Biomass Electricity Generation in Billion kWh')

# add labels and legend
ax.set_xlabel('')
ax.set_ylabel('Biomass Electricity Generation (Billion kWh)')
ax.legend(['Brazil', 'Rest of SAM'])

# display the graph
plt.show()
```
