```python
df = pd.read_csv("SeriesExport-11-12-2022_15-08-33.csv")
```

```python
df.head(10)
```

```python
column_names = list(df.columns)
filtered_columns = [name for name in column_names if not "Series Name." in name]
# filtered_columns
df = df[filtered_columns]
df.head(10)
df = df.drop(range(6)).reset_index(drop=True)
```

```python
df
```

```python
Biomass_columns = [name for name in filtered_columns if ("Biomass" in name) or ("Series Name" in name)]
# filtered_columns
# print(Biomass_columns)
Biomass_df = df[Biomass_columns]
Biomass_df = Biomass_df.dropna()
Biomass_df
```

```python
def find_non_ints(row):
    for i in range(len(row)):
#         print(row[i])
        if "(" in str(row[i]) or "ie" in str(row[i]):
#             print("here")
            row[i] = 0.0
        else:
#             print(row[i])
            row[i] = float(row[i])
            
#     print("here")
    return row
    
Biomass_df = Biomass_df.apply(lambda row: find_non_ints(row), axis=1)
Biomass_df
```

```python
list(Biomass_df.columns)[1].split(",")[1].strip()
```

```python
cleaned_column_names = [column.split(",")[1].strip() for column in list(Biomass_df.columns)[1:]]
print(["Series Name"]+cleaned_column_names)
Biomass_df.columns = ["Series Name"] + cleaned_column_names
lines = Biomass_df.plot.line(x="Series Name", title = "Biomass Electricity Generation in Billion kWh")
```
